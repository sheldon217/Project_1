```python
end_date = datetime.now().strftime("%Y-%m-%d")
Health_care_df = web.DataReader(["UNH","JNJ","MRK", "ABBV", "ABT","XLV"],start='2019-08-16',end=end_date,data_source='yahoo')['Adj Close']
```

```python
health_care_df_plot = Health_care_df.hvplot()
```

```python
health_care_daily = Health_care_df.pct_change()
health_care_daily.dropna(inplace=True)
```

```python
health_care_daily_plot = health_care_daily.hvplot()
```

```python
health_cummprod_re = (1+health_care_daily).cumprod()-1
health_cumprod_re_plot = health_cumm_re.hvplot()
```

```python
health_care_std = health_care_daily.std().sort_values(ascending=False)
```

```python
health_risk = health_care_std[health_care_std>health_care_std["XLV"]]
```
